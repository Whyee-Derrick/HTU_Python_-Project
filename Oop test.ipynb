```python
class  SmartPhone:
    def __init__(self):
        self.battery=50
        
        
    def charge(self):
        self.battery=100
        return self.battery
        
my_phone=SmartPhone()
print(f"Actual phone batttery state:{my_phone.battery}%") 

print(f"After charging:{my_phone.charge()}%")  

print(f"New battery state:{my_phone.battery}%")      
          
    
```
