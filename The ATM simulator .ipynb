```python
balance=1000
def make_deposite(current_balance,amount):
    new_balance=current_balance+amount
    return new_balance


def make_withdrawal(current_balance,amount):
    if amount>current_balance:
        print("Error : Insufficient funds")
        return current_balance
    else:
       return current_balance-amount

amount_added=float(input("Enter the amount you want to deposit"))

balance=make_deposite(balance,amount_added)
print(f"Your new balance is:${balance}")

amount_substracted=float(input("Enter the amount you want to withdral"))

balance=make_withdrawal(balance,amount_substracted)
print(f"Your new balance is:${balance}")

```
